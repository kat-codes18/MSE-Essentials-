Knowing Bragss Law  $2d\sin(\theta) =n\lambda$, write a function that calculates the interplanar spacing  given the xray wavelength and the diffraction angle. Try it for several examples to verify that it works.



```python

import numpy as np

def calculate_interplanar_spacing(wavelength, diffraction_angle, n=1):
  d = (n * wavelength) / (2 * np.sin(np.radians(diffraction_angle)))
  return d


wavelength = 1.54
angle1 = 20.0
angle2 = 35.5
angle3 = 60.0

d1 = calculate_interplanar_spacing(wavelength, angle1)
d2 = calculate_interplanar_spacing(wavelength, angle2)
d3 = calculate_interplanar_spacing(wavelength, angle3)


print(f"For angle {angle1} degrees, interplanar spacing d = {d1:.2f} Angstroms")
print(f"For angle {angle2} degrees, interplanar spacing d = {d2:.2f} Angstroms")
print(f"For angle {angle3} degrees, interplanar spacing d = {d3:.2f} Angstroms")
```

Write a function that takes in three Miller indices and a list of fractional atomic coordinates as a Nx3 array and computes a structure factor for the correponding unit cell in units of the atomic scattering factor. As a
reminder, the formula for structure factor is:


$F_{hkl} = ∑_{n=1}^{N_{unit cell}} {f_n exp(2 π i ( hu_n +kv_n + lw_n))}$


where $f_n$ is the atomic scattering factor for atom n.
Note that for more complicated functions that aren't one-liners like the earlier example above, the exercise of writing
"pseudocode" can be helpful in conceptualizing the steps you need to take. Here is some pseudocode for this function:

    F = 0;
    for each atom in unit cell
      frac_coords_here = frac_coords(atom_index, :);
      F = F + exp(2*pi*j * frac_coords_here .* hkl);

Note that to turn this into working code, you will have to add the function declaration, make sure variables get assigned properly, that for loop syntax is correct, etc. The point of pseudocode is to help you think through
the logic of the function, and once you have the "muscle memory" of how to do those other pieces, they will be comparatively easy to add in (that may not be the case yet, and that's okay! It's still good to separate the logic from the syntax, which is what pseudocode helps us do).


```python

import numpy as np

def calculate_structure_factor(hkl, frac_coords):
  structure_factor = 0
  for frac_coord in frac_coords:
    exponent = 2 * np.pi * 1j * np.dot(hkl, frac_coord)
    structure_factor += np.exp(exponent)

  return structure_factor


hkl = (1, 1, 1)
frac_coords = np.array([[0.0, 0.0, 0.0], [0.5, 0.5, 0.5]])

structure_factor = calculate_structure_factor(hkl, frac_coords)
print(f"Structure factor for hkl = {hkl}: {structure_factor}")

```


##OPTION 2 Function for Calculation of thermodynamic Mixing Energies

In thermodynamics, it is necessary to repeatedly calculate free energies to determine the solubility state  as temperature changes. Let's construct a function that uses a regular solution model for a mixture of two components. The two components will have an entropy of mixing identical to that of an ideal solution, but a non-zero enthalpy of mixing.  The free energy of mixing in such a regular solution is:

$\Delta G = \Delta H - T \Delta S$

Where ΔG is the free energy, ΔH is the enthalpy of mixing, T is temperature (in Kelvin), and ΔS is the entropy of mixing.
The enthalpy of mixing is defined as a function of the concentrations and an interaction parameter β:

$\Delta H = \beta X_aX_b$

The entropy is calculated as;

$\Delta S = -R(X_a lnX_a +X_b ln X_b)$

Here, R is the gas constant, and Xa and Xb are the concentrations of the two components of the solution, where Xa + Xb = 1.  

Write a pseudocode for a function and a script. The script will send the interaction parameter β to the function.  The value of β can range from 100 to 1000 Joules/mol.
Your function should determine if phase separation occurs in the temperature range of 0 to 1000K. Your program will calculate the free energy change  across the composition of mixture. The function should start at high temperature (1000 K) where there is little chance of a miscibility gap, then lower the temperature and look for the existence of two minima in free energy, instead of one (recall the change in free energy becomes positive when an miscibility gap occurs).  Use  a temperature resolution of 10 K.  It is not necessary, but fun to plot free energy vs. Xa or Xb at each temperature interval to determine of your function is operating properly. For each Beta, write out the value of Beta, the temperature the miscibility gap occurs at.   
Hint:   Small Beta mean the mixing is more stable, -ve Beta means the mixing is exothermic and thus A-B  bonding is preferred, and lastly +ve Beta means A-A  and B-B bonds are preferred.




```python
import numpy as np
import matplotlib.pyplot as plt

def find_minima(free_energies):
    minima = []

    for i in range(1, len(free_energies) - 1):
        if free_energies[i-1][1] > free_energies[i][1] and free_energies[i][1] < free_energies[i+1][1]:
            minima.append(free_energies[i])

    return minima


def calculate_free_energy(Xa, T, beta):
    R = 8.314
    if Xa <= 0 or Xa >= 1:
        return None
    Xb = 1 - Xa
    delta_H = beta * Xa * Xb
    delta_S = -R * (Xa * np.log(Xa) + Xb * np.log(Xb))
    delta_G = delta_H - T * delta_S

    return delta_G
temp_start = 1000
temp_end = 0
temp_step = 10
beta_values = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
xa_values = np.linspace(0.01, 0.99, 100)

for beta in beta_values:
    miscibility_gap_temp = None

for beta in beta_values:
    for T in range(temp_start, temp_end, -temp_step):
        free_energies = []

        for xa in xa_values:
            delta_G = calculate_free_energy(xa, T, beta)
            if delta_G is not None:
                free_energies.append((xa, delta_G))
        minima = find_minima(free_energies)

        if len(minima) == 2:
            miscibility_gap_temp = T
            print(f"Phase separation at T={T}K for β={beta}")
            break

    if miscibility_gap_temp is not None:
          xa_vals, G_vals = zip(*free_energies)
          plt.plot(xa_vals, G_vals, label=f'T={miscibility_gap_temp}K')
          plt.xlabel("Xa")
          plt.ylabel("Free Energy (ΔG)")
          plt.title(f"Free Energy vs Xa for β={beta} at T={miscibility_gap_temp}K")
          plt.legend()
          plt.show()
    else:
          print(f"There's no miscibility gap found in the temperature range for β={beta}.")
```
